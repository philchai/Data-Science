#### Consider the following dataframe of futures contract transactions: 
pc = pd.DataFrame({'Buy_or_Sell':['B','S','B','S','S','B'],'Price':[2204.0,2275.5,2250.75,2230.0,2280.75,2282.25],'Qty':[1,2,1,2,1,3]}).

#### Create a new column called 'Value', which is Price * Qty, but is negative if the transaction is a buy, and positive if the transaction is a sell (why?).

```python
#3a
pc['Value'] = (pc['Price']*pc['Qty']*-1).where(pc['Buy_or_Sell']=='B',pc['Qty']*pc['Price'])
pc
```

#### How would we determine the money we made on these transactions? (Assume the prices given are in dollars.)

```python
buys = pc['Value'].where(pc['Buy_or_Sell']=='B').sum()
sells = pc['Value'].where(pc['Buy_or_Sell']=='S').sum()
pnl = buys+sells
pnl
```

#### Find the average (mean) price of buy transactions.

```python
#3c
pc['Price'].where(pc['Buy_or_Sell']=='B').mean()
```
