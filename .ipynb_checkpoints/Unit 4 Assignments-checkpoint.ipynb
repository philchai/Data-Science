#### Create a vertically stacked set of two graphs using df2: the top graph will contain all of the price data together, and the bottom graph will contain the volume data. Choose different colours for each data set. (Note: you can use hex colour codes when defining colours, e.g.: color=#19F0B9. You can use the following RGB to Hex converter http://www.rgbtohex.net/)

```python
#1B
import matplotlib.gridspec as gridspec

fig = plt.figure()
# set up subplot grid; how many rows and columns
gridspec.GridSpec(2,1)

plt.subplot2grid((2,1), (0,0))  # (2,1) = total plot size, (0,0) = coordinates for *this* plot
plt.plot(df2.index,df2['open'],label='Open', color='blue')
plt.plot(df2.index,df2['high'],label='High', color='red')
plt.plot(df2.index,df2['low'],label='Low', color='green')
plt.plot(df2.index,df2['close'],label='High', color='yellow')
plt.xlabel('Time', size=12)
plt.ylabel('Price',size=12)
plt.legend()
plt.xlim(df2.index[0],df2.index[-1])
plt.title('Futures Prices',size=18)
plt.grid()
fig.autofmt_xdate()

plt.subplot2grid((2,1), (1,0))
plt.plot(df2.index,df2['volume'],label='Volume', color='blue')
plt.xlabel('Time', size=12)
plt.ylabel('Price',size=12)
plt.legend()
plt.xlim(df2.index[0],df2.index[-1])
plt.grid()
fig.autofmt_xdate()

plt.show()
```

#### Create a 3d scatter plot where the  xx -values range from  −2−2  to  44 , the  yy -values are the cosine-squared of  xx , and the  zz  values are the sine of the difference of  xx  and  yy . Set the step size for  xx  to be at most  0.010.01 .

```python
#3a
import mpl_toolkits.mplot3d as m3d
x = np.arange(-2,4,0.001)
y = np.cos(x)**2
z = np.sin(x-y)
scatter3d = m3d.Axes3D(plt.figure())
scatter3d.scatter3D(x,y,z)
plt.show()
```

#### Using the meshgrid method, graph a wireframe hyperbolic paraboloid for  a=3⎯⎯√a=3 ,  b=2b=2 . (Have the  xyxy  values range from at least  −10−10  to  1010  to see the structure properly.)

```python
#3B
X = np.arange(-10,10,0.1)
Y = np.arange(-10,10,0.1)
xx, yy = np.meshgrid(X, Y);
Z = yy**2/3 - xx**2/4
ax = m3d.Axes3D(plt.figure())
ax.plot_wireframe(xx, yy, Z, rstride=10, cstride=10);
plt.show();
```
